```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
```

```python
# another histogram 
data= ['price','horsepower','enginesize','citympg']
plt=car[data].hist(bins=5)
```

```python
# sample data
data = ['Price', 'horsepower', 'enginesize', 'citympg']

# Create a histogram
plt.hist(car.price, edgecolor='black', color='green')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
#plot 1:
plt.figure(figsize=(10, 6))
sns.countplot(x='CarBrand', data=car)
plt.title('Distribution of Car Brands')
plt.xticks(rotation=90)
plt.show()
```

```python
#plot 2: Relationship between horsepower and price 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='price', data=car)
plt.title('Relationship between Horsepower and Price')
plt.show()
```

```python
#plot 3: Distribution of fuel types
plt.figure(figsize=(10, 6))
sns.countplot(x='fueltype', data=car)
plt.title('Distribution of Fuel Types')
plt.show()
```

```python
#plot 4: comparison city MPG and highway MPG
plt.figure(figsize=(10, 6))
sns.boxplot(x='fueltype', y='citympg', data=car)
plt.title('Comparison of City MPG by Fuel Types')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='fueltype', y='highwaympg', data=car)
plt.title('comparison of Highway MPG by Fuel Type')
plt.show()
```

```python
#Get the count of each fuel type
fuel_type_counts = car['fueltype'].value_counts()

# create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Fuel Types')
plt.show()
```

```python
car_brand_counts = car['CarBrand'].value_counts().head(10) # show top 10 brands
plt.figure(figsize=(8, 8))
plt.pie(car_brand_counts, labels=car_brand_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Top 10 Car Brand')
plt.show()
```

```python
# Create a bar
plt.bar(car.fueltype,car.price,color='yellow')
plt.xlabel('fueltype')
plt.ylabel('price')
plt.title('bar')
plt.show()
```
